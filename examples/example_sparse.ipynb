# Exporting Trajectories for sparse-learning-CRN

This notebook converts trajectory data from BayesCRNInference JSON format 
into the plain-text format expected by the sparse-learning-CRN C++ tools:
https://github.com/zwpku/sparse-learning-CRN

**Workflow:**
1. Run the export cells to create working directories
2. Open a terminal and run `prepare` then `sparse_learning` from each working directory
3. Return to this notebook to read and compare results

---

## Method comparison

| | BayesCRNInference | sparse-learning-CRN |
|---|---|---|
| Kinetics assumption | Mass-action (exact) | None — fits polynomial basis functions |
| Method | Bayesian MCMC, spike-and-slab prior | L1-regularised MLE (FISTA) |
| Output | Posterior distributions | Sparse point estimates |
| Sparsity tuning | Automatic (prior) | Manual (λ) |
| Language | Python | C++ |

## Step 3 — Helper functions for reading results

```python
def read_channel_info(run_dir):
    """Read channel_info.txt and return list of stoichiometric change vectors."""
    path = os.path.join(run_dir, "output", "channel_info.txt")
    if not os.path.exists(path):
        return None
    with open(path) as f:
        lines = f.read().strip().splitlines()
    n_channels, n_species = map(int, lines[0].split())
    changes = [list(map(int, lines[i+1].split())) for i in range(n_channels)]
    return changes


def read_omega(run_dir, n_species=4, poly_order=2):
    """Read all omega_vec_for_channel_N.txt files. Returns dict: ch -> coeffs array."""
    sp = ["A", "B", "X", "Y"][:n_species]
    labels = ["1"] + sp + [sp[i]+"*"+sp[j] for i in range(n_species) for j in range(i, n_species)]
    omega = {}
    for fpath in sorted(glob.glob(os.path.join(run_dir, "output", "omega_vec_for_channel_*.txt"))):
        ch = int(os.path.basename(fpath).split("_")[-1].replace(".txt", ""))
        with open(fpath) as f:
            int(f.readline().strip())
            coeffs = np.array([float(x) for x in f.readline().split()])
        omega[ch] = coeffs
    return omega, labels


def top_terms(coeffs, labels, n=3, threshold=0.05):
    """Return top n terms by absolute value, filtered by threshold."""
    idx = np.argsort(np.abs(coeffs))[::-1]
    terms = [(labels[i], coeffs[i]) for i in idx if abs(coeffs[i]) >= threshold]
    return terms[:n]


def prop_label(r, reactant_mat, species, sampled_idx):
    """Return human-readable propensity basis function for reaction r (0-indexed in sampled set)."""
    col = sampled_idx[r]
    rv  = np.array(reactant_mat)[:, col]
    nz  = np.nonzero(rv)[0]
    sp  = species
    if len(nz) == 0: return "1"
    if len(nz) == 1:
        i, m = nz[0], int(rv[nz[0]])
        return sp[i] if m == 1 else f"{sp[i]}*({sp[i]}-1)/2"
    if len(nz) == 2:
        return sp[nz[0]] + "*" + sp[nz[1]]
    return "?"


print("Helper functions loaded.")
```

## Step 4 — Raw results: all basis coefficients

Use this cell to inspect any single run in detail.

```python
def print_all_coefficients(run_dir, threshold=1e-4):
    omega, labels = read_omega(run_dir)
    if not omega:
        print(f"No results yet in {run_dir}")
        return
    print(f"{len(omega)} channels\n")
    print(f"{'Channel':>8}  {'Basis':>12}  {'Coefficient':>14}")
    print("-" * 40)
    for ch in sorted(omega):
        for i, coef in enumerate(omega[ch]):
            if abs(coef) > threshold:
                print(f"{ch:>8}  {labels[i]:>12}  {coef:>14.6f}")

lam_str = str(FIXED_LAMBDA).replace(".", "p")

print("=== T=100 ===")
print_all_coefficients(f"../sparse_runs/example5_T100_lam{lam_str}")
print()
print("=== T=200 ===")
print_all_coefficients(f"../sparse_runs/example5_T200_lam{lam_str}")
print()
print("=== T=400 ===")
print_all_coefficients(f"../sparse_runs/example5_T400_lam{lam_str}")
```

## Step 6 — Comparison: true rates vs. sparse-learning estimates

For each channel, we show the true reactions and their propensity form, then the
recovered dominant coefficient(s) for T=100, T=200, and T=400 at the fixed lambda.

Since sparse-learning gives point estimates (not posteriors), the equivalent of the
Bayesian table is: **true rate** vs. **recovered coefficient** for the matching basis term.

```python
# Load true network
(
    reactionNetwork, CRN_stoichiometric_matrix, CRN_reaction_names,
    CRN_parameter_names, trueTheta, parameter_values, sampled_indices,
    reactant_matrix, unique_changes, compatible_reactions, species_names
) = load_reaction_network("../data/example5_network.json")

stoich = np.array(CRN_stoichiometric_matrix)   # (4, 20)
n_rxns = stoich.shape[1]

print("Network loaded:", n_rxns, "reactions,", len(species_names), "species")
```

```python
lam_str = str(FIXED_LAMBDA).replace(".", "p")

runs = {
    "T=100": f"../sparse_runs/example5_T100_lam{lam_str}",
    "T=200": f"../sparse_runs/example5_T200_lam{lam_str}",
    "T=400": f"../sparse_runs/example5_T400_lam{lam_str}",
}

# Load omega for each run (use T=200 to get channel ordering)
omega_all  = {}
channel_changes = None
for label, run_dir in runs.items():
    o, labels = read_omega(run_dir)
    omega_all[label] = o
    if o and channel_changes is None:
        channel_changes = read_channel_info(run_dir)

if channel_changes is None:
    print("No results found — run prepare and sparse_learning first.")
else:
    # Build true-reaction lookup per stoichiometric change
    def reactions_for_change(change):
        out = []
        for r in range(n_rxns):
            if list(stoich[:, r].astype(int)) == change:
                rate  = parameter_values[CRN_parameter_names[r]]
                basis = prop_label(r, reactant_matrix, species_names, sampled_indices)
                out.append((CRN_reaction_names[r].rstrip(":"), rate, basis))
        return out

    # Print comparison table
    header = (f"{'Ch':>3}  {'Stoich change':>20}  {'Reaction':>22}  "
              f"{'True rate':>10}  {'Basis':>14}  "
              + "".join(f"  {t:>12}" for t in runs.keys()))
    print(header)
    print("-" * len(header))

    for ch, change in enumerate(channel_changes):
        reactions = reactions_for_change(change)
        if not reactions:
            # No true reaction — skip or mark as noise
            rec_strs = []
            for label in runs:
                o = omega_all[label]
                if ch in o:
                    top = top_terms(o[ch], labels, n=1, threshold=0.0)
                    rec_strs.append(f"{top[0][0]}={top[0][1]:+.3f}" if top else "---")
                else:
                    rec_strs.append("---")
            print(f"{ch:>3}  {str(change):>20}  {'(noise channel)':>22}  "
                  f"{'---':>10}  {'---':>14}  " + "".join(f"  {s:>12}" for s in rec_strs))
            continue

        for r_idx, (rxn_name, true_rate, basis) in enumerate(reactions):
            # Find the recovered coefficient for this basis term
            rec_strs = []
            for label in runs:
                o = omega_all[label]
                if ch in o:
                    # Find this basis in labels
                    if basis in labels:
                        b_idx = labels.index(basis)
                        rec_strs.append(f"{o[ch][b_idx]:+.3f}")
                    else:
                        # Approximate: A*(A-1)/2 ≈ A*A/2
                        approx = basis.split("*(")[0]
                        approx_basis = approx + "*" + approx
                        if approx_basis in labels:
                            b_idx = labels.index(approx_basis)
                            rec_strs.append(f"{o[ch][b_idx]:+.3f}~")
                        else:
                            rec_strs.append("?")
                else:
                    rec_strs.append("---")

            ch_label = str(ch) if r_idx == 0 else ""
            change_label = str(change) if r_idx == 0 else ""
            print(f"{ch_label:>3}  {change_label:>20}  {rxn_name:>22}  "
                  f"{true_rate:>10.4f}  {basis:>14}  "
                  + "".join(f"  {s:>12}" for s in rec_strs))
    print()
    print("Note: '~' = coefficient is for A*A basis (approximation of A*(A-1)/2)")
```

## Step 7 — Lambda sweep comparison (T=200 only)

For each channel, show how the recovered coefficient for the **expected basis term** changes
across lambda values. Ideal: coefficient converges to true rate as lambda → 0, 
but zeroes out false positives as lambda → large.

```python
# Load omega for all sweep lambdas (T=200)
sweep_omega = {}
sweep_channel_changes = None
for lam in SWEEP_LAMBDAS:
    lam_str = str(lam).replace(".", "p")
    run_dir = f"../sparse_runs/example5_T200_lam{lam_str}"
    o, lbs = read_omega(run_dir)
    sweep_omega[lam] = o
    if o and sweep_channel_changes is None:
        sweep_channel_changes = read_channel_info(run_dir)
        sweep_labels = lbs

if sweep_channel_changes is None:
    print("No lambda sweep results found — run the T=200 sweep first.")
else:
    lam_header = "".join(f"  {lam:>8}" for lam in SWEEP_LAMBDAS)
    header2 = f"{'Ch':>3}  {'Reaction':>22}  {'True rate':>10}  {'Basis':>14}" + lam_header
    print(header2)
    print("-" * len(header2))

    for ch, change in enumerate(sweep_channel_changes):
        reactions = reactions_for_change(change)
        if not reactions:
            continue
        for r_idx, (rxn_name, true_rate, basis) in enumerate(reactions):
            rec_strs = []
            for lam in SWEEP_LAMBDAS:
                o = sweep_omega[lam]
                if ch in o and basis in sweep_labels:
                    b_idx = sweep_labels.index(basis)
                    rec_strs.append(f"{o[ch][b_idx]:+.3f}")
                elif ch in o:
                    approx = basis.split("*(")[0]
                    approx_basis = approx + "*" + approx
                    if approx_basis in sweep_labels:
                        b_idx = sweep_labels.index(approx_basis)
                        rec_strs.append(f"{o[ch][b_idx]:+.3f}~")
                    else:
                        rec_strs.append("?")
                else:
                    rec_strs.append("---")

            ch_label = str(ch) if r_idx == 0 else ""
            print(f"{ch_label:>3}  {rxn_name:>22}  {true_rate:>10.4f}  {basis:>14}  "
                  + "".join(f"  {s:>8}" for s in rec_strs))
```
