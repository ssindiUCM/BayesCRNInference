This notebook converts trajectory data from BayesCRNInference JSON format 
into the plain-text format expected by the sparse-learning-CRN C++ tools:
https://github.com/zwpku/sparse-learning-CRN

**Workflow:**
1. Run the export cells to create working directories
2. Open a terminal and run `prepare` then `sparse_learning` from each working directory
3. Return to this notebook to read and compare results

---

## Method comparison

| | BayesCRNInference | sparse-learning-CRN |
|---|---|---|
| Kinetics assumption | Mass-action (exact) | None — fits polynomial basis functions |
| Method | Bayesian MCMC, spike-and-slab prior | L1-regularised MLE (FISTA) |
| Output | Posterior distributions | Sparse point estimates |
| Sparsity tuning | Automatic (prior) | Manual (λ) |
| Language | Python | C++ |

## Step 3 — Helper functions for reading results

```python
def read_channel_info(run_dir):
    """Read channel_info.txt. Returns (changes, counts)."""
    path = os.path.join(run_dir, "output", "channel_info.txt")
    if not os.path.exists(path):
        return None, None
    with open(path) as f:
        lines = [l.strip() for l in f.read().strip().splitlines() if l.strip()]
    n_channels, n_species = map(int, lines[0].split())
    changes = [list(map(int, lines[i+1].split())) for i in range(n_channels)]
    counts  = list(map(int, lines[n_channels+1].split()))
    return changes, counts


def read_omega(run_dir, n_species=4, poly_order=2):
    """Read all omega_vec_for_channel_N.txt files. Returns dict: ch -> coeffs array."""
    sp = ["A", "B", "X", "Y"][:n_species]
    labels = ["1"] + sp + [sp[i]+"*"+sp[j] for i in range(n_species) for j in range(i, n_species)]
    omega = {}
    for fpath in sorted(glob.glob(os.path.join(run_dir, "output", "omega_vec_for_channel_*.txt"))):
        ch = int(os.path.basename(fpath).split("_")[-1].replace(".txt", ""))
        with open(fpath) as f:
            int(f.readline().strip())
            coeffs = np.array([float(x) for x in f.readline().split()])
        omega[ch] = coeffs
    return omega, labels


def top_terms(coeffs, labels, n=3, threshold=0.05):
    idx = np.argsort(np.abs(coeffs))[::-1]
    return [(labels[i], coeffs[i]) for i in idx if abs(coeffs[i]) >= threshold][:n]


def prop_label(r, reactant_mat, species, sampled_idx):
    col = sampled_idx[r]
    rv  = np.array(reactant_mat)[:, col]
    nz  = np.nonzero(rv)[0]
    sp  = species
    if len(nz) == 0: return "1"
    if len(nz) == 1:
        i, m = nz[0], int(rv[nz[0]])
        return sp[i] if m == 1 else f"{sp[i]}*({sp[i]}-1)/2"
    if len(nz) == 2:
        return sp[nz[0]] + "*" + sp[nz[1]]
    return "?"


print("Helper functions loaded.")
```

## Step 4 — Raw results: all basis coefficients

Use this cell to inspect any single run in detail.

```python
def print_all_coefficients(run_dir, threshold=1e-4):
    omega, labels = read_omega(run_dir)
    if not omega:
        print(f"No results yet in {run_dir}")
        return
    print(f"{len(omega)} channels\n")
    print(f"{'Channel':>8}  {'Basis':>12}  {'Coefficient':>14}")
    print("-" * 40)
    for ch in sorted(omega):
        for i, coef in enumerate(omega[ch]):
            if abs(coef) > threshold:
                print(f"{ch:>8}  {labels[i]:>12}  {coef:>14.6f}")

lam_str = str(FIXED_LAMBDA).replace(".", "p")

print("=== T=100 ===")
print_all_coefficients(f"../sparse_runs/example5_T100_lam{lam_str}")
print()
print("=== T=200 ===")
print_all_coefficients(f"../sparse_runs/example5_T200_lam{lam_str}")
print()
print("=== T=400 ===")
print_all_coefficients(f"../sparse_runs/example5_T400_lam{lam_str}")
```

## Step 6 — Comparison: true rates vs. sparse-learning estimates

For each channel, we show the true reactions and their propensity form, then the
recovered dominant coefficient(s) for T=100, T=200, and T=400 at the fixed lambda.

Since sparse-learning gives point estimates (not posteriors), the equivalent of the
Bayesian table is: **true rate** vs. **recovered coefficient** for the matching basis term.

```python
# Load true network
(
    reactionNetwork, CRN_stoichiometric_matrix, CRN_reaction_names,
    CRN_parameter_names, trueTheta, parameter_values, sampled_indices,
    reactant_matrix, unique_changes, compatible_reactions, species_names
) = load_reaction_network("../data/example5_network.json")

stoich = np.array(CRN_stoichiometric_matrix)   # (4, 20)
n_rxns = stoich.shape[1]

print("Network loaded:", n_rxns, "reactions,", len(species_names), "species")
```

```python
lam_str = str(FIXED_LAMBDA).replace(".", "p")

runs = {
    "T=100": f"../sparse_runs/example5_T100_lam{lam_str}",
    "T=200": f"../sparse_runs/example5_T200_lam{lam_str}",
    "T=400": f"../sparse_runs/example5_T400_lam{lam_str}",
}

omega_all       = {}
channel_changes = None
for label, run_dir in runs.items():
    o, labels = read_omega(run_dir)
    omega_all[label] = o
    if o and channel_changes is None:
        channel_changes, _ = read_channel_info(run_dir)

if channel_changes is None:
    print("No results found — run prepare and sparse_learning first.")
else:
    reactant_mat_arr = np.array(reactant_matrix)
    true_rate_by_idx = {}
    for r in range(n_rxns):
        true_rate_by_idx[sampled_indices[r]] = parameter_values[CRN_parameter_names[r]]

    def true_coeff_vector(change):
        """15-dim true polynomial coefficient vector for a channel,
        summing contributions from all active reactions with this stoich change.
        A*(A-1)/2  ->  +k/2 * A*A  -  k/2 * A  (correct stochastic expansion)."""
        c = np.zeros(len(labels))
        key = tuple(change)
        if key not in compatible_reactions:
            return c
        for full_idx in compatible_reactions[key]:
            rate = true_rate_by_idx.get(full_idx, 0.0)
            if rate == 0.0:
                continue
            rv = reactant_mat_arr[:, full_idx]
            nz = np.nonzero(rv)[0]
            sp = species_names
            if len(nz) == 0:
                c[labels.index("1")] += rate
            elif len(nz) == 1:
                i, m = nz[0], int(rv[nz[0]])
                if m == 1:
                    c[labels.index(sp[i])] += rate
                elif m == 2:
                    c[labels.index(sp[i]+"*"+sp[i])] += rate / 2.0
                    c[labels.index(sp[i])]            -= rate / 2.0
            elif len(nz) == 2:
                c[labels.index(sp[nz[0]]+"*"+sp[nz[1]])] += rate
        return c

    def poly_str(c, threshold=0.02):
        terms = [f"{v:+.3f}·{labels[i]}" for i, v in enumerate(c) if abs(v) >= threshold]
        return "  ".join(terms) if terms else "0"

    print(f"Comparison: true polynomial vs sparse-learning recovered polynomial")
    print(f"L² error computed over all {len(labels)} basis coefficients")
    print(f"(LASSO shrinkage bias: recovered coefficients are pulled toward zero)")
    print()

    for ch, change in enumerate(channel_changes):
        c_true    = true_coeff_vector(change)
        true_poly = poly_str(c_true)
        print(f"Ch {ch:2d}  {str(change)}")
        print(f"    True polynomial  : {true_poly if true_poly != '0' else '(no active reactions in candidate set)'}")
        for label in runs:
            o = omega_all[label]
            if ch not in o:
                print(f"    {label}  : (no result)")
                continue
            c_hat = o[ch]
            err   = np.linalg.norm(c_hat - c_true)
            rec   = poly_str(c_hat)
            print(f"    {label} recovered: {rec}")
            print(f"    {label} L² error : {err:.4f}")
        print()
```

## Step 7 — Lambda sweep comparison (T=200 only)

For each channel, show how the recovered coefficient for the **expected basis term** changes
across lambda values. Ideal: coefficient converges to true rate as lambda → 0, 
but zeroes out false positives as lambda → large.

```python
sweep_omega           = {}
sweep_channel_changes = None
sweep_counts          = None
sweep_labels          = None
for lam in SWEEP_LAMBDAS:
    lam_str = str(lam).replace(".", "p")
    o, lbs  = read_omega(f"../sparse_runs/example5_T200_lam{lam_str}")
    sweep_omega[lam] = o
    if o and sweep_channel_changes is None:
        sweep_channel_changes, sweep_counts = read_channel_info(
            f"../sparse_runs/example5_T200_lam{lam_str}")
        sweep_labels = lbs

if sweep_channel_changes is None:
    print("No lambda sweep results found.")
else:
    import pandas as pd

    reactant_mat_arr = np.array(reactant_matrix)
    true_rate_by_idx = {sampled_indices[r]: parameter_values[CRN_parameter_names[r]]
                        for r in range(n_rxns)}

    def true_coeff_vector_sweep(change):
        c = np.zeros(len(sweep_labels))
        key = tuple(change)
        if key not in compatible_reactions:
            return c
        for full_idx in compatible_reactions[key]:
            rate = true_rate_by_idx.get(full_idx, 0.0)
            if rate == 0.0:
                continue
            rv = reactant_mat_arr[:, full_idx]
            nz = np.nonzero(rv)[0]
            sp = species_names
            if len(nz) == 0:
                c[sweep_labels.index("1")] += rate
            elif len(nz) == 1:
                i, m = nz[0], int(rv[nz[0]])
                if m == 1:
                    c[sweep_labels.index(sp[i])] += rate
                elif m == 2:
                    c[sweep_labels.index(sp[i]+"*"+sp[i])] += rate / 2.0
                    c[sweep_labels.index(sp[i])]            -= rate / 2.0
            elif len(nz) == 2:
                c[sweep_labels.index(sp[nz[0]]+"*"+sp[nz[1]])] += rate
        return c

    rows = []
    for ch, (change, count) in enumerate(zip(sweep_channel_changes, sweep_counts)):
        c_true = true_coeff_vector_sweep(change)
        row = {"Ch": ch, "Obs": count, "Stoich change": str(change)}
        for lam in SWEEP_LAMBDAS:
            o = sweep_omega[lam]
            err = np.linalg.norm(o[ch] - c_true) if ch in o else np.nan
            row[f"λ={lam}"] = round(err, 4)
        rows.append(row)

    df = pd.DataFrame(rows).set_index("Ch")

    # Bold minimum per row (markdown doesn't support bold in tables easily,
    # so we flag it with *)
    lam_cols = [f"λ={lam}" for lam in SWEEP_LAMBDAS]
    def mark_min(row):
        vals = row[lam_cols]
        min_val = vals.min()
        return vals.apply(lambda v: f"*{v:.4f}*" if v == min_val else f"{v:.4f}")

    display_df = df[["Obs", "Stoich change"]].copy()
    for col in lam_cols:
        display_df[col] = df[col]

    print("T=200 lambda sweep — L² error per channel (bold = minimum)")
    print("True polynomial uses exact stochastic form: A*(A-1)/2 = A²/2 − A/2")
    print()
    print(display_df.to_string())
    print()
    print("* = minimum L² error for that channel")
    print("No single lambda minimises all channels simultaneously — fundamental limitation of global L1 regularisation.")
```

## Step 8 — Full polynomial coefficient table (for supplement)

For each stoichiometric channel, 4 rows are shown:
- **True** — exact stochastic polynomial coefficients (A*(A-1)/2 → A²/2 − A/2)
- **T=100, T=200, T=400** — coefficients recovered by sparse-learning at the fixed lambda

Columns are ordered to group each species' quadratic+linear pair together:
`1 | A² | A | B² | B | X² | X | Y² | Y | AB | AX | AY | BX | BY | XY`

This makes it easy to spot the coupled (k/2, −k/2) pairs that arise from bimolecular
same-species reactions, and to identify false positives (non-zero coefficients that
should be zero).

```python
lam_str = str(FIXED_LAMBDA).replace(".", "p")
T_runs = {
    "T=100": f"../sparse_runs/example5_T100_lam{lam_str}",
    "T=200": f"../sparse_runs/example5_T200_lam{lam_str}",
    "T=400": f"../sparse_runs/example5_T400_lam{lam_str}",
}

omega_T = {}
ch_changes = None
ch_counts  = None
for label, run_dir in T_runs.items():
    o, lbs = read_omega(run_dir)
    omega_T[label] = o
    if o and ch_changes is None:
        ch_changes, ch_counts = read_channel_info(run_dir)

if ch_changes is None:
    print("No results — run prepare and sparse_learning first.")
else:
    import pandas as pd

    reactant_mat_arr = np.array(reactant_matrix)
    true_rate_by_idx = {sampled_indices[r]: parameter_values[CRN_parameter_names[r]]
                        for r in range(n_rxns)}

    # Reordered basis: group quadratic+linear pairs, then cross terms
    sp = species_names  # [A, B, X, Y]
    ordered_labels = (
        ["1"] +
        [f"{s}*{s}" for s in sp] +   # A*A, B*B, X*X, Y*Y  (displayed as A², etc.)
        list(sp) +                    # A, B, X, Y
        [f"{sp[i]}*{sp[j]}" for i in range(4) for j in range(i+1, 4)]  # AB, AX, AY, BX, BY, XY
    )
    # interleave: 1, A*A, A, B*B, B, X*X, X, Y*Y, Y, then cross terms
    interleaved = ["1"]
    for s in sp:
        interleaved.append(f"{s}*{s}")
        interleaved.append(s)
    interleaved += [f"{sp[i]}*{sp[j]}" for i in range(4) for j in range(i+1, 4)]

    # Display column headers
    col_display = {
        "1": "1",
        "A*A": "A²", "A": "A",
        "B*B": "B²", "B": "B",
        "X*X": "X²", "X": "X",
        "Y*Y": "Y²", "Y": "Y",
        "A*B": "AB", "A*X": "AX", "A*Y": "AY",
        "B*X": "BX", "B*Y": "BY", "X*Y": "XY",
    }

    # Map from interleaved label -> index in original labels list
    orig_labels, _ = read_omega(list(T_runs.values())[0])
    _, orig_labels  = read_omega(list(T_runs.values())[0])

    def get_true_coeffs(change):
        c_full = np.zeros(len(orig_labels))
        key = tuple(change)
        if key not in compatible_reactions:
            return c_full
        for full_idx in compatible_reactions[key]:
            rate = true_rate_by_idx.get(full_idx, 0.0)
            if rate == 0.0:
                continue
            rv = reactant_mat_arr[:, full_idx]
            nz = np.nonzero(rv)[0]
            s  = species_names
            if len(nz) == 0:
                c_full[orig_labels.index("1")] += rate
            elif len(nz) == 1:
                i, m = nz[0], int(rv[nz[0]])
                if m == 1:
                    c_full[orig_labels.index(s[i])] += rate
                elif m == 2:
                    c_full[orig_labels.index(s[i]+"*"+s[i])] += rate / 2.0
                    c_full[orig_labels.index(s[i])]           -= rate / 2.0
            elif len(nz) == 2:
                c_full[orig_labels.index(s[nz[0]]+"*"+s[nz[1]])] += rate
        return c_full

    def extract_ordered(c_full):
        """Return coefficient array in interleaved column order."""
        return np.array([c_full[orig_labels.index(lbl)] for lbl in interleaved])

    rows = []
    for ch, (change, count) in enumerate(zip(ch_changes, ch_counts)):
        c_true    = get_true_coeffs(change)
        true_row  = extract_ordered(c_true)

        change_str = str(change)
        rows.append({"Ch": ch, "Obs": count, "Change": change_str, "Row": "True",
                     **{col_display[lbl]: round(true_row[k], 4)
                        for k, lbl in enumerate(interleaved)}})

        for t_label in T_runs:
            o = omega_T[t_label]
            if ch in o:
                rec = extract_ordered(o[ch])
            else:
                rec = np.full(len(interleaved), np.nan)
            rows.append({"Ch": ch, "Obs": "", "Change": "", "Row": t_label,
                         **{col_display[lbl]: round(rec[k], 4)
                            for k, lbl in enumerate(interleaved)}})

    df = pd.DataFrame(rows)
    df = df.set_index(["Ch", "Obs", "Change", "Row"])

    pd.set_option("display.max_columns", None)
    pd.set_option("display.width", 200)
    pd.set_option("display.float_format", lambda x: f"{x:+.3f}" if x != 0 else " 0.000")
    print(f"Polynomial coefficient table (λ={FIXED_LAMBDA}, all T values)")
    print(f"Columns: 1 | A² A | B² B | X² X | Y² Y | AB AX AY BX BY XY")
    print()
    print(df.to_string())
```
